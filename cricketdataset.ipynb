```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Train.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# c=one hot encoding
df_encoded = pd.get_dummies(df["Team1"],prefix="Team1")
df = pd.concat([df, df_encoded], axis=1)
df.drop(["Team1"], axis=1, inplace=True)
df_encoded = pd.get_dummies(df["Team2"],prefix="Team2")
df = pd.concat([df, df_encoded], axis=1)
df.drop(["Team2"], axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df.drop(["Stadium","MonthOfMatch","Team2_Innings"],axis=1,inplace=True) #dropped Innings of team2 because team1 innings are complementary, there is no new information, month of match and stadium affect wwinner negligibly
```

```python
df["Team1_Innings"]=df["Team1_Innings"].apply(lambda x: x=="First")
```

```python
df
```

```python
df["Team2_Venue"].value_counts()
```

```python
df_encoded = pd.get_dummies(df["Team1_Venue"],prefix="Team1_Venue")
df = pd.concat([df, df_encoded], axis=1)
df.drop(["Team1_Venue"], axis=1, inplace=True)

df_encoded = pd.get_dummies(df["Team2_Venue"],prefix="Team2_Venue")
df = pd.concat([df, df_encoded], axis=1)
df.drop(["Team2_Venue"], axis=1, inplace=True)

df_encoded = pd.get_dummies(df["HostCountry"],prefix="HostCountry")
df = pd.concat([df, df_encoded], axis=1)
df.drop(["HostCountry"], axis=1, inplace=True)
```

```python
df.info()
```

```python
x = np.array(df.drop(["MatchWinner"],axis=1))
y = np.array(pd.get_dummies(df["MatchWinner"]))
x_test = x[-400:]
y_test = y[-400:]
x = x[:-400]
y = y[:-400]
```

```python
x.shape
```

```python
import tensorflow as tf
import keras
from keras import layers

model = keras.Sequential(
    [
        layers.Dense(57,activation='relu'),
        layers.Dense(200,activation='relu'),
        layers.Dense(200,activation='relu'),
        layers.Dense(16,activation='sigmoid'),
    ]
)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=30)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(np.argmax(model.predict(x_test),axis=1),np.argmax(y_test,axis=1))
```
